```python
df.drop("id",axis='columns',inplace=True)
```

```python
Categorical_columns=["Gender","Have you ever had suicidal thoughts ?","Family History of Mental Illness"]
```

```python
df=pd.get_dummies(df,columns=Categorical_columns)
```

```python
df.drop(columns=["Gender_Female","Have you ever had suicidal thoughts ?_No","Family History of Mental Illness_No"],axis="columns",inplace=True)
```

```python
columns=["City","Profession","Sleep Duration","Dietary Habits","Degree"]
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder=LabelEncoder()
for column in df.columns:
    df[column]=label_encoder.fit_transform(df[column])
```

```python
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=df[[col for col in df.columns if col!='Depression']+["Depression"]]
df.head()
```

```python
x=df.iloc[:, :-1]
y=df.iloc[:, -1]
```

```python
correlation=pd.concat([x,y],axis=1)
sns.heatmap(correlation.corr(),annot=True,cmap="coolwarm")
plt.title("correlation Heatmap")
plt.show()
```

```python
sns.pairplot(df,vars=["Age","Academic Pressure","Work Pressure","CGPA","Job Satisfaction"],hue="Depression")
plt.plot()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=98,test_size=0.25,shuffle=True)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression(random_state=0)
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,y_pred)
print("The accuracy score is:",acc)
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
dt=DecisionTreeClassifier(random_state=1)
dt.fit(x_train,y_train)
```

```python
y_pred=dt.predict(x_test)
y_pred
```

```python
acc=accuracy_score(y_test,y_pred)
print("The accuracy score is:",acc)
```

## Removing Outliers

```python
sns.boxplot(x='Depression', y='Age', data=df)
plt.title('Age vs Depression')
plt.show()
```

```python
sns.boxplot(x='Depression', y='Job Satisfaction', data=df)
plt.title('Job Satisfaction vs Depression')
plt.show()
```

```python
sns.boxplot(x='Depression', y='Work Pressure', data=df)
plt.title('Study Satisfaction vs Depression')
plt.show()
```

```python
Q1=df['Age'].quantile(0.25)
Q3=df['Age'].quantile(0.75)
IQR=Q3-Q1

lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR

df1=df[(df['Age']>=lower_bound)& (df['Age']<=upper_bound)]
```

```python
Q1=df['Job Satisfaction'].quantile(0.25)
Q3=df['Job Satisfaction'].quantile(0.75)
IQR=Q3-Q1

lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR

df1=df[(df['Job Satisfaction']>=lower_bound)& (df['Job Satisfaction']<=upper_bound)]
```

```python
Q1=df['Work Pressure'].quantile(0.25)
Q3=df['Work Pressure'].quantile(0.75)
IQR=Q3-Q1

lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR

df1=df[(df['Work Pressure']>=lower_bound)& (df['Work Pressure']<=upper_bound)]
```

```python
sns.pairplot(df1,vars=["Age","Academic Pressure","Work Pressure","CGPA","Job Satisfaction"],hue="Depression")
plt.plot()
```

## Logistic Regression

```python
x=df1.iloc[:, :-1]
y=df1.iloc[:, -1]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=98,test_size=0.25,shuffle=True)
```

```python
lr=LogisticRegression(random_state=0)
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
acc=accuracy_score(y_test,y_pred)
print("Accuracy after outliers removal is:",acc)
```
